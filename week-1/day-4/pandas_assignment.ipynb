```python
#1
import pandas as pd
red_df =pd.read_csv('data/winequality-red.csv', delimiter=';')
white_df =pd.read_csv('data/winequality-white.csv', delimiter=';')
```

```python
#3
red_mean_acc=red_df['alcohol'].mean()
print(red_mean_acc)
```

```python
white_mean_acc=white_df['alcohol'].mean()
print(white_mean_acc)
```

```python
import numpy as np
red_df['acc_mean_check']=np.where(red_df['alcohol']>red_mean_acc,'Yes','No')
red_df.head()
white_df['acc_mean_check']=np.where(white_df['alcohol']>white_mean_acc,'Yes','No')
white_df.head()
```

```python
wine_can_drink=red_df[red_df['acc_mean_check']=='No'].append(white_df[white_df['acc_mean_check']=='No'])
wine_can_drink.head(10)
wine_can_drink.shape
```

```python
#4
red_bin=pd.cut(red_df['fixed acidity'], bins=5,labels=[1,2,3,4,5])
red_bin.name='cat'
new_red=[red_bin,red_df]
red_acid=pd.concat(new_red,axis=1)
red_acid.shape
print(red_acid)
```

```python
white_bin=pd.cut(white_df['fixed acidity'], bins=5,labels=[1,2,3,4,5])
white_bin.name='cat'
new_white=[white_bin,white_df]
white_acid=pd.concat(new_white,axis=1)
white_acid.shape
print(white_acid.shape)
print(white_acid)
```

```python
red_acid[red_acid['cat']<4]
white_acid[white_acid['cat']<4]
```

```python
#5
pd.pivot_table(red_df, values='alcohol', index='quality', columns='acc_mean_check',aggfunc=np.mean)
```

```python
pd.pivot_table(white_df, values='alcohol', index='quality', columns='acc_mean_check')
```

```python
#6
pd.pivot_table(red_acid, values='alcohol', index='quality', columns='cat')
```

```python
pd.pivot_table(white_acid, values='alcohol', index='quality', columns='cat')
```
